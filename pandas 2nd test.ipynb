```python
import matplotlib.pyplot as plt

try:
    s1 = data.groupby(["ship_type", "weather_conditions"])["engine_efficiency"].mean()
    df = s1.unstack()  
    df.plot(kind='bar', figsize=(10, 6), colormap='viridis')
    plt.title('Engine Efficiency by Ship Type and Weather Conditions')
    plt.xlabel('Ship Type')
    plt.ylabel('Average Engine Efficiency')
    plt.xticks(rotation=45)
    plt.legend(title="Weather Conditions")
    plt.show()

except Exception as e:
    print("Error Occurred:", str(e))
```

```python
try:
    plt.boxplot([data['fuel_consumption'], data['CO2_emissions']], labels=['fuel_consumption', 'CO2_emissions	'])
    plt.title('fuel_consumption vs CO2_emissions', fontsize=16)
    plt.ylabel('CO2_emissions')
    plt.show()
except Exception as e:
    print("Error Occurred:", str(e))
```

```python
mco2=data.groupby("month")['CO2_emissions'].sum().reset_index()
```

```python
import seaborn as sns
```

```python


fig, ax = plt.subplots(1, 1, figsize=(14, 10))  
sns.lineplot(data=mco2, x='month', y='CO2_emissions', ax=ax)  # Use 'ax' instead of axs[1,0]
ax.set_title("Month-wise CO2 Emission")  # Correct title placement
plt.show()
```

```python
##plot the regression graph for distance against fuel consumption
sns.regplot(data=data,x='fuel_consumption',y='distance',ax=ax)
ax.set_title('Regression:Distance vs Fuel Consumption')
ax.axis(False)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.regplot(data=data, x='fuel_consumption', y='distance', ax=ax)
ax.set_title('Regression: Distance vs Fuel Consumption')
ax.grid(False)  
plt.show()
```
